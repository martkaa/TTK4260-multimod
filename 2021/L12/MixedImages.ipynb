```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
im2='./data/mixed1.jpg'
im1='./data/mixed2.jpg'
```

```python
# Function to show two images
def showTwoIm(im1, im2):
    plt.figure(figsize=(28, 14))
    plt.subplot(1,2,1)
    plt.imshow(im1,cmap='gray')
    plt.axis('off')
    plt.subplot(1,2,2)
    plt.imshow(im2,cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
# Original images
S1 = cv2.imread(im1,0)
S2 = cv2.imread(im2,0)
m,n = S1.shape
showTwoIm(S1, S2)
```
